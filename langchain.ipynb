```python
from langchain.document_loaders import AzureBlobStorageContainerLoader

loaders = AzureBlobStorageContainerLoader(conn_str="DefaultEndpointsProtocol=https;AccountName=ishitagptblob1",container="shita-container-langchain")
documents = loaders.load()
print(f'You have {len(documents)} documents in your data')
```

```python
from langchain.text_splitter import CharacterTextSplitter

text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
docs = text_splitter.split_documents(documents)
```

```python
from langchain.embeddings import OpenAIEmbeddings
from langchain.llms import AzureOpenAI
from langchain.chains import RetrievalQA
from langchain.vectorstores import Chroma

embeddings = OpenAIEmbeddings(model="text-embedding-ada-002",deployment="embedding-ada-test-ishita")
```
