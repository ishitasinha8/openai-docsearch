```python
import openai
import tiktoken
import os
from langchain.retrievers import AzureCognitiveSearchRetriever
from langchain.embeddings import OpenAIEmbeddings, SentenceTransformerEmbeddings
```

```python
os.environ["AZURE_COGNITIVE_SEARCH_SERVICE_NAME"] = "acs-openai-test-3"
os.environ["AZURE_COGNITIVE_SEARCH_INDEX_NAME"] ="azureblob-index-17files"
os.environ["AZURE_COGNITIVE_SEARCH_API_KEY"] = ""

os.environ["OPENAI_API_TYPE"] = "azure"
os.environ["OPENAI_API_VERSION"] = "2023-03-15-preview"
os.environ["OPENAI_API_BASE"] = "https://testnew.openai.azure.com/"
os.environ["OPENAI_API_KEY"] = ""
```

```python
#query = input("Enter your query: ")
query = "northwind"
retriever = AzureCognitiveSearchRetriever(content_key="content")
#try:
```

```python

docs = retriever.get_relevant_documents(query, kwargs={"k":1})
print(len(docs))
print(docs)
```

```python
#Summarization
from langchain.docstore.document import Document
from langchain.chains.summarize import load_summarize_chain
from langchain import OpenAI
from langchain.text_splitter import CharacterTextSplitter

llm = OpenAI(temperature=0)

text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
docs2 = text_splitter.split_documents(docs)
print(len(docs2))
docs3 = [Document(page_content=str(t)) for t in docs2[:3]]
print(len(docs3))
chain2 = load_summarize_chain(llm, chain_type="map_reduce")
output = chain2.run(docs3)
print(output)
```

```python
from langchain.llms import OpenAI
from langchain.chains import RetrievalQA
from langchain.chains.question_answering import load_qa_chain

# qa = RetrievalQA.from_chain_type(llm=OpenAI(), chain_type="stuff", retriever=docs)
# qa.run("query") #error: Can't instantiate abstract class BaseRetriever with abstract methods _aget_relevant_documents, _get_relevant_documents (type=type_error)

chain = load_qa_chain(OpenAI(temperature=0, batch_size=2), chain_type="map_rerank")
chain.run(input_documents=docs, question=query)
#This model's maximum context length is 4097 tokens, however you requested 59938 tokens (59682 in your prompt; 256 for the completion). Please reduce your prompt; or completion length.
```

```python
from langchain.chat_models import AzureChatOpenAI
from langchain.chains.question_answering import load_qa_chain

chain = load_qa_chain(llm=AzureChatOpenAI(deployment_name="testgptturbo",model_name="gpt-35-turbo-16k", temperature=0, max_tokens=3000),chain_type="map_reduce")
chain.run(input_documents=docs3, question=query)
```
